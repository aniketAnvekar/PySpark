```python
import findspark

findspark.init()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('LinearReg').getOrCreate()
```

```python
data = spark.read.format('libsvm').option("numFeatures", "692").load('sample_linear_regression_data.txt')
data.show()
```

```python
(train_data,test_data) = data.randomSplit([0.7,0.3])
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
lr = LinearRegression(featuresCol='features',labelCol='label',predictionCol='prediction')
```

```python
lrModel = lr.fit(train_data)
```

```python
lrModel.coefficients
```

```python
lrModel.intercept
```

```python
test_predict = lrModel.evaluate(test_data)
```

```python
test_predict.predictions.show()
```

```python
test_predict.r2
```

```python
# creating unlabel data from test data

unlabel_data = test_data.select('features')
unlabel_data.show()
```

```python
predict_unlabel_data = lrModel.transform(unlabel_data)

predict_unlabel_data.show()
```

```python
spark.stop()
```
